Deep Learning with Neural Networks

This Python project demonstrates the implementation of a fundamental linear regression model using TensorFlow's user-friendly Keras API. The model is designed to learn a linear relationship between an independent variable (input feature) and a dependent variable (target value)



*   keras is a framework that will help build a nn as a sequence of layers
*   will need numpy to represent data in arrays




Screenshot 2024-11-03 215527.png



```python
import tensorflow as tf
import numpy as np
```

```python
x_array = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0])
y_array = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0])
```

Define Neural Network



*   define expected shape of i/p layer
*   one layer dense of o/p layer



```python
array_model = tf.keras.Sequential([
    tf.keras.Input(shape=(1,)),
    #adding one output layer
    tf.keras.layers.Dense(units = 1)
])
```

Compile the Neural Network


Have to specify 2 functions:   
use mean squared error for the loss and stochastic gradient descent for the optimizer
*   a loss:
measures the guessed answers against the known correct answers and measures how well or how badly it did
*   an optimizer:
to make another guess and try to minimize the loss to get closer to the correct result

repeat this for the number of epochs


```python
array_model.compile(optimizer= 'sgd', loss= 'mean_squared_error')
```

Train the model

```python
array_model.fit(x_array, y_array, epochs = 500)
```

Make prediction

```python
array_model.predict(np.array([10.0]))
```

```python
test_nn = array_model.predict(np.array([10.0]))
```

```python
print(f"model prediction for 10.0 is : {test_nn}")
```
